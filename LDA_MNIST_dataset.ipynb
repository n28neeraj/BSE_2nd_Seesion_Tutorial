#Dataset with 3 classes

```
# Prepare X and Y
Y = df.loc[:, '0']
X = df.drop(['0'], axis=1)
```

```

X_np = X.values
X_std = StandardScaler().fit_transform(X_np)
```

```
lda = LDA()
X_LDA_2D = lda.fit_transform(X_std, Y.values )
```

```
df_lda = pd.DataFrame(X_LDA_2D)
```

```
df_lda.head()
```

```
from scipy.spatial import ConvexHull

def encircle(x,y, ax=None, **kw):
    if not ax: ax=plt.gca()
    p = np.c_[x,y]
    hull = ConvexHull(p)
    poly = plt.Polygon(p[hull.vertices,:], **kw)
    ax.add_patch(poly)
```

```

fig, ax = plt.subplots(1,1, figsize=(16,12))

# Row masks for each category
rows_0 = Y==0;
rows_1 = Y==1; 
rows_2 = Y==2; 

# Plot
ax.scatter(df_lda.loc[rows_0.tolist(), 0], df_lda.loc[rows_0.tolist(), 1], c='blue', edgecolor='k', s=120, label='Zero')
ax.scatter(df_lda.loc[rows_1.tolist(), 0], df_lda.loc[rows_1.tolist(), 1], c='red', edgecolor='k', s=120, label='One')
ax.scatter(df_lda.loc[rows_2.tolist(), 0], df_lda.loc[rows_2.tolist(), 1], c='green', edgecolor='k', s=120, label='Two')

# Encircle the boundaries
encircle(df_lda.loc[rows_0.tolist(), 0], df_lda.loc[rows_0.tolist(), 1], ec="blue", fc="none", linewidth=2.5)
encircle(df_lda.loc[rows_1.tolist(), 0], df_lda.loc[rows_1.tolist(), 1], ec="firebrick", fc="none", linewidth=2.5)
encircle(df_lda.loc[rows_2.tolist(), 0], df_lda.loc[rows_2.tolist(), 1], ec="green", fc="none", linewidth=2.5)

# Shading
encircle(df_lda.loc[rows_1.tolist(), 0], df_lda.loc[rows_1.tolist(), 1], ec="k", fc="firebrick", alpha=0.05)
encircle(df_lda.loc[rows_0.tolist(), 0], df_lda.loc[rows_0.tolist(), 1], ec="k", fc="blue", alpha=0.05)
encircle(df_lda.loc[rows_2.tolist(), 0], df_lda.loc[rows_2.tolist(), 1], ec="k", fc="green", alpha=0.05)

# Labels
ax.set_title("MNIST Data for 0, 1 & 2", fontsize=22)
ax.set_xlabel("1st Linear Discriminant", fontsize=22)
ax.set_ylabel("2nd Linear Discriminant", fontsize=22)
ax.legend(loc='best', fontsize=16)
plt.show();
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
np_pca = pca.fit_transform(X=X_np)
```

```
df_pca=pd.DataFrame(np_pca)
fig, ax = plt.subplots(1,1, figsize=(16,12))

# Row masks for each category
rows_0 = Y==0;
rows_1 = Y==1; 
rows_2 = Y==2; 

# Plot
ax.scatter(df_pca.loc[rows_0.tolist(), 0], df_pca.loc[rows_0.tolist(), 1], c='blue', edgecolor='k', s=120, label='Zero')
ax.scatter(df_pca.loc[rows_1.tolist(), 0], df_pca.loc[rows_1.tolist(), 1], c='red', edgecolor='k', s=120, label='One')
ax.scatter(df_pca.loc[rows_2.tolist(), 0], df_pca.loc[rows_2.tolist(), 1], c='green', edgecolor='k', s=120, label='Two')

# Encircle the boundaries
encircle(df_pca.loc[rows_0.tolist(), 0], df_pca.loc[rows_0.tolist(), 1], ec="blue", fc="none", linewidth=2.5)
encircle(df_pca.loc[rows_1.tolist(), 0], df_pca.loc[rows_1.tolist(), 1], ec="firebrick", fc="none", linewidth=2.5)
encircle(df_pca.loc[rows_2.tolist(), 0], df_pca.loc[rows_2.tolist(), 1], ec="green", fc="none", linewidth=2.5)

# Shading
encircle(df_pca.loc[rows_1.tolist(), 0], df_pca.loc[rows_1.tolist(), 1], ec="k", fc="firebrick", alpha=0.05)
encircle(df_pca.loc[rows_0.tolist(), 0], df_pca.loc[rows_0.tolist(), 1], ec="k", fc="blue", alpha=0.05)
encircle(df_pca.loc[rows_2.tolist(), 0], df_pca.loc[rows_2.tolist(), 1], ec="k", fc="green", alpha=0.05)

# Labels
ax.set_title("MNIST Data for 0, 1 & 2", fontsize=22)
ax.set_xlabel("1st principal component", fontsize=22)
ax.set_ylabel("2nd principal component", fontsize=22)
ax.legend(loc='best', fontsize=16)
plt.show();
```
